```python
import pandas as pd
import dill as pickle
import numpy as np
```

## Reading and flattening data into a single data frame

```python
with open("eu-news-batch.pkl", "rb") as file:
    data_raw = pickle.load(file)
```

```python
# Flatten the nested dictionary
flattened_data = []

for country, journals in data_raw.items():
    for journal, pillars in journals.items():
        for pillar, languages in pillars.items():
            for lang, articles in languages.items():
              for article in articles:
                try:
                    flattened_data.append({
                        "country"   : country,
                        "journal"   : journal,
                        "asspillar" : pillar,
                        "language"  : lang,
                        **article
                    })
                except TypeError:
                  print(f"Issues found for articles in {country}, {journal}, and {lang}")
```

```python
master_data = pd.DataFrame(flattened_data)
master_data.drop_duplicates(subset = "article_id", keep = "first", inplace = True)
```

## Wrangling the data

```python
lang_codes = {
    "german"    : "de",
    "finnish"   : "fi",
    "swedish"   : "sv",
    "croatian"  : "hr",
    "polish"    : "pl",
    "french"    : "fr",
    "spanish"   : "es",
    "portuguese": "pt",
    "italian"   : "it",
    "greek"     : "el",
    "dutch"     : "nl",
    "hungarian" : "hu",
    "bulgarian" : "bg",
    "estonian"  : "et",
    "romanian"  : "ro",
    "slovak"    : "sk",
    "danish"    : "da",
    "slovenian" : "sl",
    "lithuanian": "lt",
    "latvian"   : "lv",
    "english"   : "en"
}
master_data["language_id"] = master_data["language"].map(lang_codes)
```

```python
compilers = ["carlos", "santi", "dalia", "artha", "other"]
master_data["compiler"] = master_data["article_id"].transform(lambda x: np.random.choice(compilers, replace = True))
```

```python
master_data.to_parquet("../data/eu-news-batch-unique.parquet.gzip", compression = "gzip")
```
